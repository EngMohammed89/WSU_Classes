Load . mat file and plot given FRFs for review.

```matlab
load mckiddat.mat
```

There are a total of 3 FRFs found in the given data. From the graphs it appears that it is a 3 DOF system.

```matlab
%Plot for column #1

plot(Freq, 20*log10(abs(Recep(:,1))), 'r','linewidth', 2)
title('H_{11} FRF');
xlabel('Frequency, Hz');
ylabel('Mag (dB)');
grid on
```

```matlab
% Plot for column #2

plot(Freq, 20*log10(abs(Recep(:,2))), 'r','linewidth', 2)
title('H_{12} FRF');
xlabel('Frequency, Hz');
ylabel('Mag (dB)');
grid on
```

```matlab
% Plot for column #3

plot(Freq, 20*log10(abs(Recep(:,3))), 'r','linewidth', 2)
title('H_{13} FRF');
xlabel('Frequency, Hz');
ylabel('Mag (dB)');
grid on
```
