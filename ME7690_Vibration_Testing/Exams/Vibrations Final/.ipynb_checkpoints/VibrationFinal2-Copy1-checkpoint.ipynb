### Define required values from the FRF graphs. This will be used to estimate the mode shapes of the system.

To determine mode shapes basaed on the estimated damping ratios above the following equation is used.  

$H_{m,l} = \Sigma_{i=1}^n\frac{_{i}A_{m,l}}{2\zeta_i j}$ where following assumptions were made regarding $\tilde{h_i}$ at the peak locations:  

$\tilde{h_i} = \frac{\omega_i^2}{\omega_i^2 + 2\zeta_i\omega_i\omega j - \omega^2} = \frac{1}{2\zeta_i j}$

```python
H11P1 = 61.48
H11P2 = 24.97
H11P3 = 2.98

H12P1 = 49.26
H12P2 = 9.71
H12P3 = 3.68

H13P1 = 19.81
H13P2 = 16.35
H13P3 = 2.58

z1 = 0.0029
z2 = 0.0033
z3 = 0.0082
```

Get following values:

- $^{1}\psi_{1}$
- $^{2}\psi_{1}$
- $^{3}\psi_{1}$

```python
phi1_11 = sp.sqrt(H11P1/(2*z1))
phi2_11 = sp.sqrt(H11P2/(2*z2))
phi3_11 = sp.sqrt(H11P3/(2*z3))
display(phi1_11)
display(phi2_11)
display(phi3_11)
```

Get following values:

- $^{1}\psi_{2}$
- $^{2}\psi_{2}$
- $^{3}\psi_{2}$

```python
phi1_12 = (H12P1/(2*z1))/phi1_11
phi2_12 = (H12P2/(2*z2))/phi2_11
phi3_12 = (H12P3/(2*z3))/phi3_11
display(phi1_12)
display(phi2_12)
display(phi3_12)
```

Get following values:

- $^{1}\psi_{3}$
- $^{2}\psi_{3}$
- $^{3}\psi_{3}$

```python
phi1_13 = (H13P1/(2*z1))/phi1_11
phi2_13 = (H13P2/(2*z2))/phi2_11
phi3_13 = (H13P3/(2*z3))/phi3_11
display(phi1_13)
display(phi2_13)
display(phi3_13)
```

**Construct the vectors and normalize**

```python
phi1 = sp.array([phi1_11, phi1_12, phi1_13])
phi1 = phi1/sp.linalg.norm(phi1)
phi1
```

```python
phi2 = sp.array([phi2_11, phi2_12, phi2_13])
phi2 = phi2/sp.linalg.norm(phi2)
phi2
```

```python
phi3 = sp.array([phi3_11, phi3_12, phi3_13])
phi3 = phi3/sp.linalg.norm(phi3)
phi3
```
