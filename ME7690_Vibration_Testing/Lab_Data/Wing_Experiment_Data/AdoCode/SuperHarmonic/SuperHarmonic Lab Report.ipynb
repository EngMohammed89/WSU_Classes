### Introduction:  

During the wing lab a possible nonlinearity was observed in the response of the accelerometers when the wing was excited at $25.85$ Hz. After the wing was allowed to reach a steady state reponse, the FRF display on the Bobcat system indicated 2 peaks active. One of the peaks was at the driving frequency of $25.85$ Hz. The other peak occured at the $2^{nd}$ natural frequency of $51.7$ Hz. Since this corresponds to exactly 2 times the driving frequency it was determined that the wing may indeed posses a super-harmonic reponse. At this point it is unclear if the non-linear response is due to the connecting rod between the wing and the shaker or if this property is independednt of the connecting rod shape and length.

It should also be noted the manner in which steady state was defined. Before any data was recorded it was ensured that the first natural frequency was completly gone from the FRF response of the wing. The main reason for this is the fact that the higer modes will die out before the first mode does. Therefore, if the first mode is not registeded on the scope, it is a good indicator that the response has indeed reached a steady point in the response.  

There were a total of 3 accelerometers attached to the wing with the following nomenclature:  

- **Channel 1: accelerometer at the end of the wing.**
- **Channel 2: accelerometer in the middle of the wing.**
- **Channel 4: accelerometer closest to the base of the wing.**
- **Channel 3: load cell measurement for the shaker.**

A total of 5 data samples 3.2 seconds long were recorded. All the driving frequencies were suppose to be at $25.85$ but when the data was evaluated it was discovered that data samples 2 and 3 had a different driving frequency, which was $28.75$ Hz. At this time it is not clear how this happened but this error can be clearly seen in the frequency plots of the input and output channels.
