$$\begin{bmatrix}
\dot{x_1} \\
\dot{x_2} \\
\end{bmatrix}
=\begin{bmatrix}
0   &   1\\
-1  &   0\\
\end{bmatrix}
*\begin{bmatrix}
x_1      \\
x_2      \\
\end{bmatrix}
+\begin{bmatrix}
0        \\

\end{bmatrix}
$$

```python
%pylab inline
import numpy as np
import scipy as sp
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def solve_2(max_time=1.0, x0 = 1, v0 = 1, IN1 = 'x2', IN2 = '-x1-x1**3'):
    
    def prob_deriv(x1_x2, t):
        
        x1, x2= x1_x2
        return [eval(IN1), eval(IN2)]
    
    #Join initial conditions into one matrix with 2 columns
    x0 = np.concatenate((x0, v0), axis = 1)
    
    #Set N to number of rows in x0
    N = x0.shape[0]
    
    # Solve for the trajectories
    t = np.linspace(0, max_time, int(250*max_time))
    
    x_t = np.asarray([sp.integrate.odeint(prob_deriv, x0i, t)
                      for x0i in x0])

    for i in range(N):
        x, v = x_t[i,:,:].T            
        plt.plot(x, v,'-')
        #Let's plot '*' at the end of each trajectory.
        plt.plot(x[-1],v[-1],'*')
        plt.grid('on')
    
    return t, x_t
```

### Problem 3:  

Determine when following equations are dissipative and conservative  

$\dot{x}=\sigma(y-x),  \dot{y}=\rho x-y-xz,  \dot{z}=-\beta z+xy$ in state space the equations becomes:  

$$\begin{bmatrix}
\dot{x}        \\
\dot{y}        \\
\dot{z}        \\
\end{bmatrix}
=\begin{bmatrix}
-\sigma  &  \sigma  &  0   \\
\rho     &  -1      &  -x  \\
0        &   0      &  -\beta  \\
\end{bmatrix}
*\begin{bmatrix}
x      \\
y      \\
z      \\
\end{bmatrix}
+\begin{bmatrix}
0        \\
0        \\
xy       \\
\end{bmatrix}
$$  
To determine when system is conservative or dissipative need to calculate $\nabla \cdot f$, which yields $-\sigma-1-\beta$. Change in volume is expressed as $V'(t) = (\nabla \cdot f)*V_o$, which when solved gives:  
<br>
**$V(t) = V_o*e^{-(\sigma+1+\beta)*t}$.**  
<br>
Based in this equation if the term $-(\sigma+1+\beta) = 0$ the volume at some time t will be equal to initial volume $V_o$. If on the other hand $-(\sigma + 1 + \beta)<0$ the system will be dissipative.

```python
%pylab inline
#%matplotlib qt
import numpy as np
import scipy as sp
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x0 = np.array([[1]])
y0 = np.array([[1]])
z0 = np.array([[1]])

def solve_3(max_time=1.0, sigma=1.0, rho=1.0, beta=1.0, x0=x0, y0=y0, z0=z0,\
            IN1 = 'sigma*(y-x)', IN2 = 'rho*x-y-x*z', IN3 = '-beta*z+x*y'):
    
    def prob_deriv(x_y_z, t, sigma=sigma, rho=rho, beta=beta):
        
        x, y, z= x_y_z
        return [eval(IN1), eval(IN2), eval(IN3)]
    
    #Join initial conditions into one matrix with 2 columns
    x00 = np.concatenate((x0, y0, z0), axis = 1)
    
    #Set N to number of rows in x0
    N = x0.shape[0]
    
    # Solve for the trajectories
    t = np.linspace(0, max_time, int(250*max_time))
    
    x_t = np.asarray([sp.integrate.odeint(prob_deriv, x00i, t)
                      for x00i in x00])
    
    Lables=['X', 'Y', 'Z']
    
    for i in range(N):
        for j in range(0, 3):
            #plot phase plots
            if j==2:
                fig=plt.figure(3)
                plt.plot(x_t[i,:,0], x_t[i,:,2],'-')
                #Let's plot '*' at the end of each trajectory.
                plt.plot(x_t[:,len(t)-1,0],x_t[:,len(t)-1,2],'*')
                #Plot box
                plt.plot(x_t[:,0,0],x_t[:,0,2],'k-')
                plt.plot(x_t[:,len(t)-1,0],x_t[:,len(t)-1,2],'k-')
                plt.xlabel(Lables[0])
                plt.ylabel(Lables[2])
            else:
                fig=plt.figure(j+1)
                plt.plot(x_t[i,:,j], x_t[i,:,j+1],'-')
                #Let's plot '*' at the end of each trajectory.
                plt.plot(x_t[:,len(t)-1,j],x_t[:,len(t)-1,j+1],'*')
                #Plot box
                plt.plot(x_t[:,0,j],x_t[:,0,j+1],'k-')
                plt.plot(x_t[:,len(t)-1,j],x_t[:,len(t)-1,j+1],'k-')
                plt.xlabel(Lables[j] + '-' + str(j))
                plt.ylabel(Lables[j+1] + '-' + str(j))
            plt.grid('on')
    
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    ax.plot(x_t[0,:,0], x_t[0,:,1], x_t[0,:,2], 'b-')
    ax.plot(x_t[1,:,0], x_t[1,:,1], x_t[1,:,2], 'r-')
    ax.plot(x_t[2,:,0], x_t[2,:,1], x_t[2,:,2], 'g-')
    ax.plot(x_t[3,:,0], x_t[3,:,1], x_t[3,:,2], 'y-')
    
    ax.plot(x_t[:,0,0], x_t[:,0,1], x_t[:,0,2], 'k-')
    ax.plot(x_t[:,len(t)-1,0], x_t[:,len(t)-1,1], x_t[:,len(t)-1,2], 'k-')
    
    #Calculate initial Area
    Vect1 = np.array([[x_t[0,0,0], x_t[0,0,1], x_t[0,0,2]]])
    Vect2 = np.array([[x_t[1,0,0], x_t[1,0,1], x_t[1,0,2]]])
    Vect3 = np.array([[x_t[2,0,0], x_t[2,0,1], x_t[2,0,2]]])
    Vect4 = np.array([[x_t[3,0,0], x_t[3,0,1], x_t[3,0,2]]])
    
    Vect12 = Vect1 - Vect2
    Vect13 = Vect1 - Vect3
    Vect14 = Vect1 - Vect4
    Vect23 = Vect2 - Vect3
    Vect24 = Vect2 - Vect4
    Vect34 = Vect3 - Vect4
    
    h1 = (np.linalg.norm(Vect12))
    h2 = (np.linalg.norm(Vect13))
    h3 = (np.linalg.norm(Vect14))
    h4 = (np.linalg.norm(Vect23))
    h5 = (np.linalg.norm(Vect24))
    h6 = (np.linalg.norm(Vect34))
    
    Ainit = h1*h3
    
    #Calculate final Area
    Vect1f = np.array([[x_t[0,len(t)-1,0], x_t[0,len(t)-1,1], x_t[0,len(t)-1,2]]])
    Vect2f = np.array([[x_t[1,len(t)-1,0], x_t[1,len(t)-1,1], x_t[1,len(t)-1,2]]])
    Vect3f = np.array([[x_t[2,len(t)-1,0], x_t[2,len(t)-1,1], x_t[2,len(t)-1,2]]])
    Vect4f = np.array([[x_t[3,len(t)-1,0], x_t[3,len(t)-1,1], x_t[3,len(t)-1,2]]])
    
    Vect12f = Vect1f - Vect2f
    Vect13f = Vect1f - Vect3f
    Vect14f = Vect1f - Vect4f
    Vect23f = Vect2f - Vect3f
    Vect24f = Vect2f - Vect4f
    Vect34f = Vect3f - Vect4f
    
    h1f = (np.linalg.norm(Vect12f))
    h2f = (np.linalg.norm(Vect13f))
    h3f = (np.linalg.norm(Vect14f))
    h4f = (np.linalg.norm(Vect23f))
    h5f = (np.linalg.norm(Vect24f))
    h6f = (np.linalg.norm(Vect34f))
    
    Afin = (h2f*(0.5*h5f)) 
    
    print('Initial area = ' + str(round(Ainit,4)) + ' and final area = '\
          + str(round(Afin,4)) + ' for the 3D plot')
    
    plt.xlabel('X');
    plt.ylabel('Y');
    ax.set_zlabel('Z');
    
    plt.show()
    
    return t, x_t
```

### Problem 5:  

Determine when following map is dissipative.  

$x_{n+1} = a_{11}x_n + a_{12}y_n$  
$y_{n+1} = a_{21}x_n + a_{22}y_n$  

The Jacobian matrix of the above map is:  
<br>
$$
\begin{bmatrix}
a_{11}  &  a_{12}  \\
a_{21}  &  a_{22}  \\
\end{bmatrix}
$$
<br>
The determinant of the above matrix needs to be $<0$ in order to have a dissopative system:  
<br>
$a_{12}a_{22} - a_{12}a_{21}<0$

```python
def prob5(a11 = 0.2, a12 = 0.2, a21=0.2, a22=0.2, firstindex = 0, \
                numsteps = 10, x0 = 0.0, y0 = 0.0):
    if firstindex > 0:
        firstindex = 0
    if -firstindex > numsteps:
        numsteps = -firstindex
    numsteps = numsteps + 1
    fig = plt.figure()
    ax = fig.add_axes([0, 0, 1, 1])
    x = np.zeros(numsteps)
    y = np.zeros(numsteps)
        
    x[-firstindex] = x0
    y[-firstindex] = y0

    def Fun_F(x, y, a11 = a11, a12 = a12, a21 = a21, a22 = a22):
        return [a11*x + a12*y, a21*x + a22*y]
    
    for i in range(-firstindex + 1, numsteps):
        x[i], y[i] = Fun_F(x[i-1], y[i-1], a11, a12, a21, a22)
  
    plt.plot(x,y,'*')
    
    for i in range(numsteps):
        #ax.annotate('%s' % i, xy=[x[i],y[i]], textcoords='offset points')
        ax.annotate('{}'.format(i + firstindex), xy=(x[i],y[i]), xytext=(3, -14), ha='right',textcoords='offset points')
    
    xrange = max(x) - min(x)
    yrange = max(y) - min(y)
    buf = 0.1
    ax.set_xlim((min(x) - buf * xrange, max(x) + buf * xrange))
    ax.set_ylim((min(y) - buf * yrange, max(y) + buf * yrange))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.show()
```

```python
prob5(a11 = 0.1, a12 = 0.5, a21=0.5, a22=0.3, firstindex = 0, \
      numsteps = 40, x0 = 1.0, y0 = 0.0)
```
