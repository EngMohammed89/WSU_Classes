## Problem 3.17  

Determine max and min of the below objective function.

<img src="Prob3_17.png" width="400">  

The inequality constraint $x^2 + y^2 + 2x \ge16$ is the general form of the equation of the circle. It is beneficial to put this equation into center radius form, which is done by completing the square. The revised inequality constraint takes the following form.  

$$(x + 1)^2 + y^2 \ge17$$  

This form is used to graph the constraint.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pylab

def c_func317(f=20, lin_den=100, axis_span=[-1, 5, -1, 5],\
           Sol_Space=[-6,6], test_val=0, cust_loc=[5, 15]):

    fig=plt.figure()
    
    x=np.linspace(-20.0,20.0,500)
    
    #circle=plt.Circle((0,0), radius=1.5, color='b', fill=False)
    circ=plt.Circle((-1,0), radius=8.5, color='r', fill=False, linewidth=4)
    fig = plt.gcf()
    fig.gca().add_artist(circ)
    
    plt.annotate("$g_1$",xy=(-1,1),xytext=(7,-5), fontsize=f)
                 #,arrowprops=dict(facecolor='black'), fontsize=f)
       
    #Plot annotations
    plt.axis(axis_span)
    fig.suptitle('Countour of $f(x_1, x_2)$', fontsize=f)
    plt.xlabel('$x$', fontsize=f)
    plt.ylabel('$y$', fontsize=f)
    plt.grid()
    
    #Contour plot
    y1=np.linspace(Sol_Space[0], Sol_Space[1],200)
    y2=np.linspace(Sol_Space[0], Sol_Space[1],200)
    X, Y = np.meshgrid(y1, y2)
    F=9*X**2 + 13*Y**2 + 18*X*Y - 4
    CS=plt.contour(X, Y, F, lin_den)
    if test_val==1:
        manual_locations = [(cust_loc)]
        plt.clabel(CS, inline=1, fontsize=f, manual=manual_locations)
    else:
        plt.clabel(CS, inline=0, fontsize=10)
```

```python
c_func317(f=20, lin_den=50, axis_span=[-13, 11, -10, 10],\
       Sol_Space=[-15,15], test_val=0, cust_loc=[5, 15])
```

From the above plot it appears that the minimum function value at about 1400.00 while the max function value is infinite. Zooming in onto the min point actually shows that the min value is approximately 1200.00.

Constraint $g_1$ is active at the minimum point.

```python
c_func317(f=20, lin_den=7, axis_span=[2.5, 6.5, 4, 10],\
       Sol_Space=[-1,10], test_val=1, cust_loc=[4.5, 6])
```

## Problem 3.21  

<img src="Prob3_21.png" width="500">  

Above problem has following standard form:  

**Find:**

- $\vec{x} = [b=width, d=depth]$  

**To minimize cross-sec area:**  

- $A = b*d$  

**Subject to inequality constraints:**

- Bending stress constraint: $\frac{6M}{bd^2}\le\sigma_a$
- Shear stress constraint: $\frac{3V}{2bd}\le\tau_a$
- Aspect ratio constraint: $d-2b\le0$  

**Side constraints:**  

- $b, d\ge0$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pylab

def c_func321(f=20, lin_den=100, axis_span=[-1, 5, -1, 5],\
           Sol_Space=[-6,6], test_val=0, cust_loc=[5, 15]):

    M=1.4e7
    V=2.4e4
    oa=1.65e4
    ta=5000
    
    fig=plt.figure()
    
    x=np.linspace(0.005,70.0,500)
    
    plt.plot((6*M)/(oa*x**2), x,'g-',label = '$g_1$', linewidth=4)
    plt.annotate("$g_1$",xy=(0.4,3.6),xytext=(9,25), fontsize=f)
    
    #plt.annotate("$g_1$",xy=(0.4,3.6),xytext=(0.25,2.8),\
                 #arrowprops=dict(facecolor='black'), fontsize=f)
    
    plt.plot((3*V)/(2*x*ta), x,'y-',label = '$g_2$', linewidth=4)
    plt.annotate("$g_2$",xy=(2,0.5),xytext=(1,15), fontsize=f)
    
    plt.plot(0.5*x, x,'b-',label = '$g_2$', linewidth=4)
    plt.annotate("$g_3$",xy=(2,0.5),xytext=(6,10), fontsize=f)

    plt.plot(x,x*0,'r-',label = '$g_2$', linewidth=4)
    plt.annotate("$g_4$",xy=(2,0.5),xytext=(4,-2.5), fontsize=f)
    
    plt.plot(x*0,x,'r-',label = '$g_2$', linewidth=4)
    plt.annotate("$g_5$",xy=(2,0.5),xytext=(-1.5,10), fontsize=f)
    
    plt.annotate("$Design Space$",xy=(2,0.5),xytext=(7.5,7.5), fontsize=f-3)
    
    #Plot annotations
    plt.axis(axis_span)
    fig.suptitle('Countour of $f(x_1, x_2)$', fontsize=f)
    plt.xlabel('$b$', fontsize=f)
    plt.ylabel('$d$', fontsize=f)
    plt.grid()
    
    #Contour plot
    y1=np.linspace(Sol_Space[0], Sol_Space[1],200)
    y2=np.linspace(Sol_Space[0], Sol_Space[1],200)
    X, Y = np.meshgrid(y1, y2)
    F=X*Y
    CS=plt.contour(X, Y, F, lin_den)
    if test_val==1:
        manual_locations = [(cust_loc)]
        plt.clabel(CS, inline=1, fontsize=f, manual=manual_locations)
    else:
        plt.clabel(CS, inline=0, fontsize=10)
```

```python
c_func321(f=20, lin_den=20, axis_span=[-2, 20, -4, 40],\
           Sol_Space=[-1,40], test_val=0, cust_loc=[5, 15])
```

From the above plot it appears that the minimum function value at about 250. Zooming in onto the min point actually shows that the min value is approximately 232 cm$^2$.

Optimum depth d $\approx$ 21.7  
Optimum width b $\approx$ 10.8

Constraints $g_1$ and $g_3$ are active at the minimum point.

```python
c_func321(f=20, lin_den=150, axis_span=[7.5, 13, 18, 25],\
           Sol_Space=[-1,30], test_val=1, cust_loc=[11.5, 20])
```

## Problem 3.27  

<img src="Prob3_27.png" width="500">  

Above problem has following standard form:  

**Find:**

- $\vec{x} = [h=height, A=area]$  

**To minimize:**  

- $Cost = 0.6h + 0.001A$  

**Subject to inequality constraints:**

- $g_1$ Floor space: $\frac{hA}{3.5}\ge20,000$
- $g_2$ Shear stress constraint: $(10,000 - A)\ge 0.25\frac{hA}{3.5}$

**Side constraints:**  

- $3.5\le h\le 21$, m
- $0\le A \le 10,000$, m$^2$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pylab

def c_func327(f=20, lin_den=100, axis_span=[-1, 5, -1, 5],\
           Sol_Space=[-6,6], test_val=0, cust_loc=[5, 15], labels=1):
    
    fig=plt.figure()
    
    x=np.linspace(0.001,12000.0,10000)
    x1=np.ones(10000, dtype=np.int)
    
    plt.plot((20000*3.5)/x, x, 'g-',label = '$g_1$', linewidth=4)
   
    plt.plot(140000/(x+14), x, 'y-',label = '$g_2$', linewidth=4)
    
    plt.plot(x, 3.5*x1,'r-',label = '$g_2$', linewidth=4)
    
    plt.plot(x, 21*x1,'r-',label = '$g_2$', linewidth=4)
    
    plt.plot(10000*x1, x,'b-',label = '$g_2$', linewidth=4)
    
    plt.plot(x*0, x, 'b-',label = '$g_2$', linewidth=4)

    if labels == 1:
        plt.annotate("$g_1$",xy=(0.4,3.6),xytext=(8000,10), fontsize=f)
        plt.annotate("$g_2$",xy=(2,0.5),xytext=(8000,5), fontsize=f)
        plt.annotate("$g_3$",xy=(2,0.5),xytext=(2000,5), fontsize=f)
        plt.annotate("$g_4$",xy=(2,0.5),xytext=(2000,22.5), fontsize=f)
        plt.annotate("$g_6$",xy=(2,0.5),xytext=(10300,10), fontsize=f)
        plt.annotate("$g_5$",xy=(2,0.5),xytext=(500,10), fontsize=f)
        plt.annotate("$Design Space$",xy=(2,0.5),xytext=(6000,15), fontsize=f-3)
    
    #Plot annotations
    plt.axis(axis_span)
    fig.suptitle('Countour of $f(x_1, x_2)$', fontsize=f)
    plt.xlabel('$A$', fontsize=f)
    plt.ylabel('$h$', fontsize=f)
    plt.grid()
    
    #Contour plot
    y1=np.linspace(Sol_Space[0], Sol_Space[1],2000)
    y2=np.linspace(Sol_Space[0], 40,500)
    X, Y = np.meshgrid(y1, y2)
    F=0.6*Y + 0.001*X
    CS=plt.contour(X, Y, F, lin_den)
    if test_val==1:
        manual_locations = [(cust_loc)]
        plt.clabel(CS, inline=1, fontsize=f, manual=manual_locations)
    else:
        plt.clabel(CS, inline=0, fontsize=10)
```

```python
c_func327(f=20, lin_den=20, axis_span=[-100, 11000, -1, 25],\
           Sol_Space=[-1,11000], test_val=0, cust_loc=[5, 15], labels=1)
```

From the above plot it appears that the min cost is about 14 million. Zooming in onto the min point actually shows that the min cost is approximately 13.5 million.

## Problem 3.37  

<img src="Prob3_37.png" width="500">  

Above problem has following standard form:  

**Find:**

- $\vec{x} = [h=height, r=radius]$  

**To maximize:**  

- $Volume = \pi r^2 h$  

**Subject to inequality constraints:**

- $g_1$ Surface area: $2\pi rh + \pi r^2 -900 \le 0$

**Side constraints:**  

- $5 \le r \le 20$, cm
- $0 \le h \le 20$, cm

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pylab

def c_func337(f=20, lin_den=100, axis_span=[-1, 5, -1, 5],\
           Sol_Space=[-6, 6, -6, 6], test_val=0, cust_loc=[5, 15], labels=1):
    
    fig=plt.figure()
    
    x=np.linspace(0.001,30,500)
    x1=np.ones(500, dtype=np.int)
    
    plt.plot(x, -((np.pi*x**2 - 900)/(2*np.pi*x)), 'g-',label = '$g_1$', linewidth=4)
    
    plt.plot(x1*5, x,'r-',label = '$g_2$', linewidth=4)
    
    plt.plot(x1*20, x,'r-',label = '$g_2$', linewidth=4)
    
    plt.plot(x, x*0,'b-',label = '$g_2$', linewidth=4)
    
    plt.plot(x, x1*20, 'b-',label = '$g_2$', linewidth=4)

    if labels == 1:
        plt.annotate("$g_1$",xy=(0.4,3.6),xytext=(10,15), fontsize=f)
        plt.annotate("$g_2$",xy=(2,0.5),xytext=(2.5,10), fontsize=f)
        plt.annotate("$g_3$",xy=(2,0.5),xytext=(21,10), fontsize=f)
        plt.annotate("$g_4$",xy=(2,0.5),xytext=(25,2), fontsize=f)
        #plt.annotate("$g_6$",xy=(2,0.5),xytext=(10,10), fontsize=f)
        plt.annotate("$g_5$",xy=(2,0.5),xytext=(25,22), fontsize=f)
        plt.annotate("$Design Space$",xy=(2,0.5),xytext=(5,2), fontsize=f-3)
    
    #Plot annotations
    plt.axis(axis_span)
    fig.suptitle('Countour of $f(x_1, x_2)$', fontsize=f)
    plt.xlabel('$r$', fontsize=f)
    plt.ylabel('$h$', fontsize=f)
    plt.grid()
    
    #Contour plot
    y1=np.linspace(Sol_Space[0], Sol_Space[1], 500)
    y2=np.linspace(Sol_Space[2], Sol_Space[3], 500)
    X, Y = np.meshgrid(y1, y2)
    F=np.pi*X**2*Y
    CS=plt.contour(X, Y, F, lin_den)
    if test_val==1:
        manual_locations = [(cust_loc)]
        plt.clabel(CS, inline=1, fontsize=f, manual=manual_locations)
    else:
        plt.clabel(CS, inline=0, fontsize=10)
```

From the above plot it appears that the max volume is about 2000 cm$^3$. Zooming in onto the max point actually shows that the max volume is approximately 2400 cm$^3$.

## Problem 3.38  

<img src="Prob3_38.png" width="500">  

Above problem has following standard form:  

**Find:**

- $\vec{x} = [n=number\ of\ tubes, r=radius]$  

**To maximize:**  

- $Area = 2\pi r l n$  

**Subject to inequality constraints:**

- $g_1$ Min radius: $r - 0.5 \ge 0$
- $g_2$ Max X-sec Area: $\pi r^2 n - 2000 \le 0$

**Side constraints:**  

- $n \ge 0$, cm

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pylab

def c_func338(f=20, lin_den=100, axis_span=[-1, 5, -1, 5],\
           Sol_Space=[-6, 6, -6, 6], test_val=0, cust_loc=[5, 15], labels=1):
    
    fig=plt.figure()
    
    x=np.linspace(0.001,2600,2600)
    x1=np.ones(2600, dtype=np.int)
    
    plt.plot(x, np.sqrt((2000)/(np.pi*x)), 'g-',label = '$g_1$', linewidth=4)
    
    plt.plot(x, x1*0.5,'r-',label = '$g_2$', linewidth=4)
    
    plt.plot(x1*0, x,'r-',label = '$g_2$', linewidth=4)
    
    plt.plot(x1, x1*0,'b-',label = '$g_2$', linewidth=4)

    if labels == 1:
        plt.annotate("$g_1$",xy=(0.4,3.6),xytext=(250,2.2), fontsize=f)
        plt.annotate("$g_2$",xy=(2,0.5),xytext=(1000,-0.25), fontsize=f)
        plt.annotate("$g_3$",xy=(2,0.5),xytext=(-150,4), fontsize=f)
        #plt.annotate("$g_4$",xy=(2,0.5),xytext=(25,2), fontsize=f)
        #plt.annotate("$g_6$",xy=(2,0.5),xytext=(10,10), fontsize=f)
        #plt.annotate("$g_5$",xy=(2,0.5),xytext=(25,22), fontsize=f)
        plt.annotate("$Design Space$",xy=(2,0.5),xytext=(0,1), fontsize=f-5)
    
    #Plot annotations
    plt.axis(axis_span)
    fig.suptitle('Countour of $f(x_1, x_2)$', fontsize=f)
    plt.xlabel('$n$', fontsize=f)
    plt.ylabel('$r$', fontsize=f)
    plt.grid()
    
    #Contour plot
    y1=np.linspace(Sol_Space[0], Sol_Space[1], 500)
    y2=np.linspace(Sol_Space[2], Sol_Space[3], 500)
    X, Y = np.meshgrid(y1, y2)
    F= 2*X*Y*np.pi*1
    CS=plt.contour(X, Y, F, lin_den)
    if test_val==1:
        manual_locations = [(cust_loc)]
        plt.clabel(CS, inline=1, fontsize=f, manual=manual_locations)
    else:
        plt.clabel(CS, inline=0, fontsize=10)
```

```python
c_func338(f=20, lin_den=50, axis_span=[-150, 2600, -1, 10],\
           Sol_Space=[0, 2600, 0, 30], test_val=0, cust_loc=[5, 15], labels=1)
```

## Problem 3.45  

<img src="Prob3_45.png" width="500">  

Above problem has following standard form:  

**Find:**

- $\vec{x} = [d=diameter, h=height]$  

**To minimize:**  

- $Area = \pi dh + 2(\frac{\pi}{4}d^2)$  

**Subject to inequality constraints:**

- $g_1$ required volume: $\frac{\pi}{4}d^2h - 400 \ge 0$

**Side constraints:**  

- $3.5 \le d \le 8$, cm
- $8 \le h \le 18$, cm

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pylab

def c_func345(f=20, lin_den=100, axis_span=[-1, 5, -1, 5],\
           Sol_Space=[-6, 6, -6, 6], test_val=0, cust_loc=[5, 15], labels=1):
    
    fig=plt.figure()
    
    x=np.linspace(0.001,100,500)
    x1=np.ones(500, dtype=np.int)
    
    plt.plot(x, np.sqrt((1600)/(np.pi*x)), 'g-',label = '$g_1$', linewidth=4)
    
    plt.plot(x, x1*3.5,'b-',label = '$g_2$', linewidth=4)
    
    plt.plot(x, x1*8,'b-',label = '$g_2$', linewidth=4)
    
    plt.plot(x1*8, x,'r-',label = '$g_2$', linewidth=4)
    
    plt.plot(x1*18, x,'r-',label = '$g_2$', linewidth=4)

    if labels == 1:
        plt.annotate("$g_1$", xy=(0.4,3.6), xytext=(10, 6), fontsize=f)
        plt.annotate("$g_2$", xy=(0.4, 3.6), xytext=(5, 8.5), fontsize=f)
        plt.annotate("$g_2$", xy=(0.4, 3.6), xytext=(5, 4), fontsize=f)
        plt.annotate("$g_3$", xy=(0.4, 3.6), xytext=(6.5, 6), fontsize=f)
        plt.annotate("$g_3$", xy=(0.4, 3.6), xytext=(19, 7), fontsize=f)
        plt.annotate("$Design Space$", xy=(0.4, 3.6), xytext=(11, 7), fontsize=f-4)
        
    #Plot annotations
    plt.axis(axis_span)
    fig.suptitle('Countour of $f(x_1, x_2)$', fontsize=f)
    plt.xlabel('$h$', fontsize=f)
    plt.ylabel('$d$', fontsize=f)
    plt.grid()
    
    #Contour plot
    y1=np.linspace(Sol_Space[0], Sol_Space[1], 500)
    y2=np.linspace(Sol_Space[2], Sol_Space[3], 500)
    X, Y = np.meshgrid(y1, y2)
    F= np.pi*X*Y + 2*Y**2*((np.pi)/(4))
    CS=plt.contour(X, Y, F, lin_den)
    if test_val==1:
        manual_locations = [(cust_loc)]
        plt.clabel(CS, inline=1, fontsize=f, manual=manual_locations)
    else:
        plt.clabel(CS, inline=0, fontsize=10)
```
