## Problem 3.1

<img src="Prob3-1.png" width="400">  

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pylab

def c_func(f=20, lin_den=100, axis_span=[-1, 5, -1, 5],\
           Sol_Space=[-6,6], test_val=0, cust_loc=[5, 15]):

    fig=plt.figure()
    
    x=np.linspace(-20.0,20.0,500)
    
    plt.plot(x,4-x,'g-',label = '$g_1$', linewidth=4)
    plt.annotate("$g_1$",xy=(0.4,3.6),xytext=(0.25,2.8),\
                 arrowprops=dict(facecolor='black'), fontsize=f)
    
    plt.plot(x,x-x,'r-',label = '$g_2$', linewidth=4)
    plt.annotate("$g_2$",xy=(2,0.5),xytext=(2,-0.3), fontsize=f)
    
    plt.plot(x-x,x,'r-',label = '$g_2$', linewidth=4)
    plt.annotate("$g_3$",xy=(2,0.5),xytext=(-0.35,2.0), fontsize=f)
    
    plt.annotate("$Design Space$",xy=(2,0.5),xytext=(0.5,1), fontsize=f-3)
    
    #Plot annotations
    plt.axis(axis_span)
    fig.suptitle('Countour of $f(x_1, x_2)$', fontsize=f)
    plt.xlabel('$x_1$', fontsize=f)
    plt.ylabel('$x_2$', fontsize=f)
    plt.grid()
    
    #Contour plot
    y1=np.linspace(Sol_Space[0], Sol_Space[1],200)
    y2=np.linspace(Sol_Space[0], Sol_Space[1],200)
    Y1, Y2 = np.meshgrid(y1, y2)
    F=(Y1-3)**2+(Y2-3)**2
    CS=plt.contour(Y1, Y2, F, lin_den)
    if test_val==1:
        manual_locations = [(cust_loc)]
        plt.clabel(CS, inline=1, fontsize=f, manual=manual_locations)
    else:
        plt.clabel(CS, inline=0, fontsize=10)
```

```python
c_func(f=20, lin_den=100, axis_span=[-1, 5, -1, 5],\
       Sol_Space=[-6,6], test_val=0, cust_loc=[5, 15])
```

From the above plot it appears that the minimum function value is at 2.0. Figure below zooms in at the minimum point to confirm that 2.0 is in fact the optimum value. The objective function is non-linear in this case but the constraints are linear

Constraint $g_1$ is active at the optimum point.

```python
c_func(f=20, lin_den=100, axis_span=[1.75, 2.5, 1.75, 2.25],\
       Sol_Space=[0,3], test_val=1, cust_loc=[2, 2])
```

## Problem 3.3

<img src="Prob3-3.png" width="400">  

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pylab

def c_func33(f=20, lin_den=5, axis_span=[-1, 5, -1, 5],\
             Sol_Space=[-6,6], test_val=0, cust_loc=[5, 15]):

    fig=plt.figure()
    
    x=np.linspace(-20.0,20.0,500)
    
    plt.plot(48-4*x,x,'g-',label = '$g_1$', linewidth=4)
    plt.annotate("$g_1$",xy=(5,10.4),xytext=(4,9),\
                 arrowprops=dict(facecolor='black'), fontsize=f)

    plt.plot(x,50-5*x,'y-',label = '$g_1$', linewidth=4)
    plt.annotate("$g_2$",xy=(8.5,7),xytext=(7,6),\
                 arrowprops=dict(facecolor='black'), fontsize=f)
    
    plt.plot(x,x-x,'r-',label = '$g_2$', linewidth=4)
    plt.annotate("$g_3$",xy=(2,0.5),xytext=(2,0.4), fontsize=f)
    
    plt.plot(x-x,x,'r-',label = '$g_2$', linewidth=4)
    plt.annotate("$g_4$",xy=(2,0.5),xytext=(0.3,2.0), fontsize=f)
    
    plt.annotate("$Design Space$",xy=(2,0.5),xytext=(2.5,4), fontsize=f-3)
    
    #Plot annotations
    plt.axis(axis_span)
    fig.suptitle('Countour of $f(x_1, x_2)$', fontsize=f)
    plt.xlabel('$x_1$', fontsize=f)
    plt.ylabel('$x_2$', fontsize=f)
    plt.grid()
    
    #Contour plot
    y1=np.linspace(Sol_Space[0], Sol_Space[1],200)
    y2=np.linspace(Sol_Space[0], Sol_Space[1],200)
    Y1, Y2 = np.meshgrid(y1, y2)
    F=Y1+3*Y2
    CS=plt.contour(Y1, Y2, F, lin_den)
    if test_val==1:
        manual_locations = [(cust_loc)]
        plt.clabel(CS, inline=1, fontsize=f, manual=manual_locations)
    else:
        plt.clabel(CS, inline=0, fontsize=10)
```

```python
c_func33(f=20, lin_den=20, axis_span=[-0.5, 11, -0.5, 13],\
             Sol_Space=[-1,15], test_val=0, cust_loc=[5, 15])
```

From the above plot it appears that the minimum function value is at approximately 0.0. Figure below zooms in at the maximum point to confirm that 0.0 is in fact the optimum value. Both the constraints and objective function are linear.

Side constraints $g_3$ and $g_4$ are active at the optimum point.

```python
c_func33(f=20, lin_den=50, axis_span=[-1, 1, -1, 1],\
             Sol_Space=[-3,3], test_val=1, cust_loc=[-0.5, 0.25])
```

## Problem 3.5

<img src="Prob3-5.png" width="400">  

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pylab

def c_func35(f=20, lin_den=5, axis_span=[-1, 5, -1, 5],\
             Sol_Space=[-6,6], test_val=0, cust_loc=[5, 15]):

    fig=plt.figure()
    
    x=np.linspace(-20.0,20.0,500)
    
    plt.plot(x,20-x,'g-',label = '$g_1$', linewidth=4)
    plt.annotate("$g_1$",xy=(8,12),xytext=(7,9.5),\
                 arrowprops=dict(facecolor='black'), fontsize=f)

    plt.plot(x,10+x,'y-',label = '$g_1$', linewidth=4)
    plt.annotate("$g_2$",xy=(2.0,12.0),xytext=(2.5,10),\
                 arrowprops=dict(facecolor='black'), fontsize=f)
    
    plt.plot(x,x-x,'r-',label = '$g_2$', linewidth=4)
    plt.annotate("$g_3$",xy=(2,0.5),xytext=(5,0.75), fontsize=f)
    
    plt.plot(x-x,x,'r-',label = '$g_2$', linewidth=4)
    plt.annotate("$g_4$",xy=(2,0.5),xytext=(0.3,5.0), fontsize=f)
    
    plt.annotate("$Design Space$",xy=(2,0.5),xytext=(2.5,4), fontsize=f-3)
    
    #Plot annotations
    plt.axis(axis_span)
    fig.suptitle('Countour of $f(x_1, x_2)$', fontsize=f)
    plt.xlabel('$x_1$', fontsize=f)
    plt.ylabel('$x_2$', fontsize=f)
    plt.grid()
    
    #Contour plot
    y1=np.linspace(Sol_Space[0], Sol_Space[1],200)
    y2=np.linspace(Sol_Space[0], Sol_Space[1],200)
    Y1, Y2 = np.meshgrid(y1, y2)
    F=4*Y1*Y2
    CS=plt.contour(Y1, Y2, F, lin_den)
    
    if test_val==1:
        manual_locations = [(cust_loc)]
        plt.clabel(CS, inline=1, fontsize=f, manual=manual_locations)
    else:
        plt.clabel(CS, inline=0, fontsize=10)
```

```python
c_func35(f=20, lin_den=20, axis_span=[-1, 20, -1, 20],\
         Sol_Space=[-1,20], test_val=0, cust_loc=[5, 15])
```

From the above plot it appears that the maximum function value is at approximately 300.0. Figure below zooms in at the maximum point to confirm that 300.0 is in fact the optimum value. Constraints are linear while the objective function is nonlinear.

Constraints $g_1$ and $g_2$ are active at the optimum point.

```python
c_func35(f=20, lin_den=40, axis_span=[3, 7, 13, 17],\
         Sol_Space=[-1,20], test_val=1, cust_loc=[5, 15])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pylab

def c_func37(f=20, lin_den=5, axis_span=[-1, 5, -1, 5],\
             Sol_Space=[-6,6], test_val=0, cust_loc=[5, 15]):

    fig=plt.figure()
    
    x=np.linspace(-20.0,20.0,500)
    
    plt.plot(x,(21-2*x)/4,'g-',label = '$g_1$', linewidth=4)
    plt.annotate("$g_1$",xy=(6.1,2.3),xytext=(7,3),\
                 arrowprops=dict(facecolor='black'), fontsize=f)

    plt.plot(x,(18-5*x)/3,'y-',label = '$g_1$', linewidth=4)
    plt.annotate("$g_2$",xy=(2.6,2.0),xytext=(4,2),\
                 arrowprops=dict(facecolor='black'), fontsize=f)
    
    plt.plot(x,x-x,'r-',label = '$g_2$', linewidth=4)
    plt.annotate("$g_3$",xy=(2,0.5),xytext=(5,0.4), fontsize=f)
    
    plt.plot(x-x,x,'r-',label = '$g_2$', linewidth=4)
    plt.annotate("$g_4$",xy=(2,0.5),xytext=(-0.8,2.0), fontsize=f)
    
    plt.annotate("$Design Space$",xy=(2,0.5),xytext=(0,0.5), fontsize=f-5)
    
    #Plot annotations
    plt.axis(axis_span)
    fig.suptitle('Countour of $f(x_1, x_2)$', fontsize=f)
    plt.xlabel('$x_1$', fontsize=f)
    plt.ylabel('$x_2$', fontsize=f)
    plt.grid()
    
    #Contour plot
    y1=np.linspace(Sol_Space[0], Sol_Space[1],200)
    y2=np.linspace(Sol_Space[0], Sol_Space[1],200)
    Y1, Y2 = np.meshgrid(y1, y2)
    F=3*Y1+Y2
    CS=plt.contour(Y1, Y2, F, lin_den)
    
    if test_val==1:
        manual_locations = [(cust_loc)]
        plt.clabel(CS, inline=1, fontsize=f, manual=manual_locations)
    else:
        plt.clabel(CS, inline=0, fontsize=10)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pylab

def c_func39(f=20, lin_den=5, axis_span=[-1, 5, -1, 5],\
             Sol_Space=[-6,6], test_val=0, cust_loc=[5, 15]):

    fig=plt.figure()
    
    x=np.linspace(-20.0,20.0,500)
    
    plt.plot(-x**2,x,'g-',label = '$g_1$', linewidth=4)
    plt.annotate("$g_1$",xy=(-1,1),xytext=(-1.2,1.2), fontsize=f)
                 #,arrowprops=dict(facecolor='black'), fontsize=f)

    #circle=plt.Circle((0,0), radius=1.5, color='b', fill=False)
    circ=plt.Circle((0,0), radius=3, color='r', fill=False, linewidth=4)
    fig = plt.gcf()
    fig.gca().add_artist(circ)
    
    plt.annotate("$g_2$",xy=(2.6,2.0),xytext=(2.2,2.2), fontsize=f)
                 #,arrowprops=dict(facecolor='black'), fontsize=f)
    
    plt.annotate("$Design Space$",xy=(2,0.5),xytext=(-2.7,0.0), fontsize=f-5)
    
    #Plot annotations
    plt.axis(axis_span)
    fig.suptitle('Countour of $f(x_1, x_2)$', fontsize=f)
    plt.xlabel('$x_1$', fontsize=f)
    plt.ylabel('$x_2$', fontsize=f)
    plt.grid()
    
    #Contour plot
    y1=np.linspace(Sol_Space[0], Sol_Space[1],200)
    y2=np.linspace(Sol_Space[0], Sol_Space[1],200)
    Y1, Y2 = np.meshgrid(y1, y2)
    F=Y1*Y2
    CS=plt.contour(Y1, Y2, F, lin_den)
    
    if test_val==1:
        manual_locations = [(cust_loc)]
        plt.clabel(CS, inline=1, fontsize=f, manual=manual_locations)
    else:
        plt.clabel(CS, inline=0, fontsize=12)
```

```python
c_func39(f=20, lin_den=150, axis_span=[-5, 3.2, -3.2, 3.2],\
             Sol_Space=[-13,13], test_val=0, cust_loc=[5, 15])
```

From the above plot it appears that the minimum function value is at approximately 0.0. Figure below zooms in at the minimum point to confirm that 0.0 is the optimum value. Constraints and the objective function are nonlinear.
