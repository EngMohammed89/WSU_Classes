First point of the discussion will be the results using numerical optimization algorithm called "fmincon", which is a function found in matlab optimization toolbox.

Function "fmincon" is used for constrained optimization problems and results for problem 3.34 can be found below.

Next, numerical solution is confirmed using the graphical method, which agrees quite well.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pylab

def c_func71(f=20, lin_den=100, axis_span=[-1, 150, -1, 2],\
           Sol_Space=[-1, 150, -1, 2], test_val=0, cust_loc=[5, 15], labels=1):
    
    fig=plt.figure()
    
    x=np.linspace(0, 500, 5000)
    x2=np.linspace(0.6, 0.999, 500)
    x1=np.ones(5000, dtype=np.int)
    
    plt.plot(x, (1.0 - (185200)/(x**3))**(1/4), 'g-', label = '$g_1$', linewidth=4)
    
    plt.plot(x, (1.0 - (18237783.112)/(x**4))**(1/4), 'b-', label = '$g_2$', linewidth=4)
          
    plt.plot(((479.3335)/((-x2+1)**(5/2)))**(1/3), x2,'y-',label = '$g_2$', linewidth=4)
    
    plt.plot(x1*20, x,'r-',label = '$g_2$', linewidth=4)
    
    plt.plot(x1*500, x,'r-',label = '$g_2$', linewidth=4)
    
    plt.plot(x, x1*0.6,'r-',label = '$g_2$', linewidth=4)
    
    plt.plot(x, x1*0.999,'r-',label = '$g_2$', linewidth=4)

    if labels == 1:
        plt.annotate("$g_1$", xy=(0.4,3.6), xytext=(200, 0.8), fontsize=f)
        #plt.annotate("$g_2$", xy=(0.4, 3.6), xytext=(5, 8.5), fontsize=f)
        #plt.annotate("$g_2$", xy=(0.4, 3.6), xytext=(5, 4), fontsize=f)
        #plt.annotate("$g_3$", xy=(0.4, 3.6), xytext=(6.5, 6), fontsize=f)
        #plt.annotate("$g_3$", xy=(0.4, 3.6), xytext=(19, 7), fontsize=f)
        #plt.annotate("$Design Space$", xy=(0.4, 3.6), xytext=(11, 7), fontsize=f-4)
        
    #Plot annotations
    plt.axis(axis_span)
    fig.suptitle('Countour of $f(x_1, x_2)$', fontsize=f)
    plt.xlabel('$x_1$', fontsize=f)
    plt.ylabel('$x_2$', fontsize=f)
    plt.grid()
    
    #Contour plot
    y1=np.linspace(Sol_Space[0], Sol_Space[1], 1000)
    y2=np.linspace(Sol_Space[2], Sol_Space[3], 1000)
    X, Y = np.meshgrid(y1, y2)
    F = 3.08269e-3*X**2*(1-Y**2)
    CS=plt.contour(X, Y, F, lin_den)
    if test_val==1:
        manual_locations = [(cust_loc)]
        plt.clabel(CS, inline=1, fontsize=f, manual=manual_locations)
    else:
        plt.clabel(CS, inline=0, fontsize=10)
```

```python
import warnings
warnings.filterwarnings('ignore')

c_func71(f=20, lin_den=50, axis_span=[10, 525, 0.5, 1.1],\
           Sol_Space=[-1, 525, -1, 2], test_val=0, cust_loc=[5, 15], labels=1)

#to turn on warnings do following
warnings.filterwarnings('default')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pylab

def c_func75(f=20, lin_den=100, axis_span=[-1, 300, -1, 30],\
           Sol_Space=[-1, 300, -1, 30], test_val=0, cust_loc=[5, 15], labels=1):
    
    with open("g1_data.txt") as f:
        data = f.read()
        
    data = data.split('\n')
    a = [row.split(' ')[0] for row in data]
    b = [row.split(' ')[1] for row in data]
        
    with open("g4_data.txt") as f1:
        data1 = f1.read()
        
    data1 = data1.split('\n')
    a1 = [row.split(' ')[0] for row in data1]
    b1 = [row.split(' ')[1] for row in data1]    
    
    fig=plt.figure()
    
    x=np.linspace(0,532,5000)
    x1=np.ones(5000, dtype=np.int)
    
    plt.plot(a, b, 'g-', label = '$g_1$', linewidth=4)
    
    plt.plot(x, -2*x + 70, 'r-', label = '$g_2$', linewidth=4)
    
    plt.plot(x, (2*x)/(91),'y-',label = '$g_2$', linewidth=4)
    
    plt.plot(a1, b1,'b-',label = '$g_2$', linewidth=4)
    
    plt.plot(x, 2*x-500,'r-',label = '$g_2$', linewidth=4)
    
    plt.plot(x, -2*x + 70,'r-',label = '$g_2$', linewidth=4)
    
    plt.plot(x, x1*40,'r-',label = '$g_2$', linewidth=4)
    
    plt.plot(x, x1*1,'r-',label = '$g_2$', linewidth=4)

    if labels == 1:
        plt.annotate("$g_1$", xy=(0.4,3.6), xytext=(200, 0.8), fontsize=f)
        #plt.annotate("$g_2$", xy=(0.4, 3.6), xytext=(5, 8.5), fontsize=f)
        #plt.annotate("$g_2$", xy=(0.4, 3.6), xytext=(5, 4), fontsize=f)
        #plt.annotate("$g_3$", xy=(0.4, 3.6), xytext=(6.5, 6), fontsize=f)
        #plt.annotate("$g_3$", xy=(0.4, 3.6), xytext=(19, 7), fontsize=f)
        #plt.annotate("$Design Space$", xy=(0.4, 3.6), xytext=(11, 7), fontsize=f-4)
        
    #Plot annotations
    plt.axis(axis_span)
    fig.suptitle('Countour of $f(x_1, x_2)$', fontsize=20)
    plt.xlabel('$R$', fontsize=20)
    plt.ylabel('$t$', fontsize=20)
    plt.grid()
    
    #Contour plot
    y1=np.linspace(Sol_Space[0], Sol_Space[1], 1000)
    y2=np.linspace(Sol_Space[2], Sol_Space[3], 1000)
    X, Y = np.meshgrid(y1, y2)
    F = 153.71*X*Y
    CS=plt.contour(X, Y, F, lin_den)
    if test_val==1:
        manual_locations = [(cust_loc)]
        plt.clabel(CS, inline=1, fontsize=20, manual=manual_locations)
    else:
        plt.clabel(CS, inline=0, fontsize=10)
```

```python
import warnings
warnings.filterwarnings('ignore')

c_func75(f=20, lin_den=20, axis_span=[-1, 300, -1, 45],\
           Sol_Space=[-1, 300, -1, 45], test_val=0, cust_loc=[5, 15], labels=0)

#to turn on warnings do following
warnings.filterwarnings('default')
```

**Looking at the plot above, optimum solution is around 56,000, which agrees with the numerical solution of 57,000. From the plot it can be seen that constraints g1 and g3 are active.**

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pylab

def c_func78(f=20, lin_den=100, axis_span=[50, 500, 0.5, 50],\
           Sol_Space=[-1, 300, -1, 30], test_val=0, cust_loc=[5, 15], labels=1):   
    
    fig=plt.figure()
    
    x=np.linspace(48,502,5000)
    x2=np.linspace(0.5,50,5000)
    x1=np.ones(5000, dtype=np.int)
    
    plt.plot(x, ((1.69765/x)*(x**2+4800)**(1/2))**(1/2), 'g-', label = '$g_1$', linewidth=4)
    
    plt.plot(x, (((0.01100852390)/(x))*(x**2+4800)**(3/2))**(1/4), 'b-', label = '$g_2$', linewidth=4)
        
    plt.plot(x1*50, x2,'r-',label = '$g_2$', linewidth=4)
    
    plt.plot(x1*500, x2,'r-',label = '$g_2$', linewidth=4)
    
    plt.plot(x, x1*0.5,'r-',label = '$g_2$', linewidth=4)
    
    plt.plot(x, x1*50,'r-',label = '$g_2$', linewidth=4)

    if labels == 1:
        plt.annotate("$g_1$", xy=(0.4,3.6), xytext=(200, 0.8), fontsize=f)
        #plt.annotate("$g_2$", xy=(0.4, 3.6), xytext=(5, 8.5), fontsize=f)
        #plt.annotate("$g_2$", xy=(0.4, 3.6), xytext=(5, 4), fontsize=f)
        #plt.annotate("$g_3$", xy=(0.4, 3.6), xytext=(6.5, 6), fontsize=f)
        #plt.annotate("$g_3$", xy=(0.4, 3.6), xytext=(19, 7), fontsize=f)
        #plt.annotate("$Design Space$", xy=(0.4, 3.6), xytext=(11, 7), fontsize=f-4)
        
    #Plot annotations
    plt.axis(axis_span)
    fig.suptitle('Countour of $f(x_1, x_2)$', fontsize=20)
    plt.xlabel('$H$', fontsize=20)
    plt.ylabel('$D$', fontsize=20)
    plt.grid()
    
    #Contour plot
    y1=np.linspace(Sol_Space[0], Sol_Space[1], 1000)
    y2=np.linspace(Sol_Space[2], Sol_Space[3], 1000)
    X, Y = np.meshgrid(y1, y2)
    F = 6.59734e-3*Y**2*(X**2 + 4800)**(1/2)
    CS=plt.contour(X, Y, F, lin_den)
    if test_val==1:
        manual_locations = [(cust_loc)]
        plt.clabel(CS, inline=1, fontsize=20, manual=manual_locations)
    else:
        plt.clabel(CS, inline=0, fontsize=10)
```

### Graphical Solution to problem 2 (part d)

```python
def c_func2(f=20, lin_den=100, axis_span=[50, 500, 0.5, 50],\
           Sol_Space=[-1, 300, -1, 30], test_val=0, cust_loc=[5, 15], labels=1):   
    
    fig=plt.figure()
    
    x=np.linspace(-50,50,500)
    x2=np.linspace(0.5,50,500)
    x1=np.ones(500, dtype=np.int)
    
    plt.plot(x, -2*x + 4, 'g-', label = '$g_1$', linewidth=4)
    
    plt.plot(x, -x - 5, 'b-', label = '$g_2$', linewidth=4)
        
    plt.plot(-x1*5, x,'r-',label = '$g_2$', linewidth=4)
    
    plt.plot(x, x1*5,'r-',label = '$g_2$', linewidth=4)

    if labels == 1:
        plt.annotate("$g_1$", xy=(0.4,3.6), xytext=(200, 0.8), fontsize=f)
        #plt.annotate("$g_2$", xy=(0.4, 3.6), xytext=(5, 8.5), fontsize=f)
        #plt.annotate("$g_2$", xy=(0.4, 3.6), xytext=(5, 4), fontsize=f)
        #plt.annotate("$g_3$", xy=(0.4, 3.6), xytext=(6.5, 6), fontsize=f)
        #plt.annotate("$g_3$", xy=(0.4, 3.6), xytext=(19, 7), fontsize=f)
        #plt.annotate("$Design Space$", xy=(0.4, 3.6), xytext=(11, 7), fontsize=f-4)
        
    #Plot annotations
    plt.axis(axis_span)
    fig.suptitle('Countour of $f(x_1, x_2)$', fontsize=20)
    plt.xlabel('$x_1$', fontsize=20)
    plt.ylabel('$x_2$', fontsize=20)
    plt.grid()
    
    #Contour plot
    y1=np.linspace(Sol_Space[0], Sol_Space[1], 1000)
    y2=np.linspace(Sol_Space[2], Sol_Space[3], 1000)
    X, Y = np.meshgrid(y1, y2)
    F = X**2 + 10*Y**2 - 3*X*Y
    CS=plt.contour(X, Y, F, lin_den)
    if test_val==1:
        manual_locations = [(cust_loc)]
        plt.clabel(CS, inline=1, fontsize=20, manual=manual_locations)
    else:
        plt.clabel(CS, inline=0, fontsize=10)
```
